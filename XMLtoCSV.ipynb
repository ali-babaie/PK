```python
import csv
import xml.etree.ElementTree as ET

# parse xml file
tree = ET.parse('2012-09-11_15_55_21.xml')
root = tree.getroot()

# open a file for writing
spaces_data = open('spaces.csv', 'w')

# create the csv writer object
csvwriter = csv.writer(spaces_data)
csv_headers = ['id','occupied','center_x','center_y', 'size_w', 'size_h', 'angel', 'point_x_1', 'point_y_1', 
                 'point_x_2', 'point_y_2', 'point_x_3', 'point_y_3', 'point_x_4', 'point_y_4']
csvwriter.writerow(csv_headers)

# loop through xml
for space in root:
    row = []
    row.append(space.attrib['id'])
    if 'occupied'in space.attrib:
        row.append(space.attrib['occupied'])
    else:
        row.append('0')
    rotatedRect = space.find('rotatedRect')
    # center
    center = rotatedRect.find('center').attrib
    row = row + [center['x'], center['y']]
    # size
    size = rotatedRect.find('size').attrib
    row = row + [size['w'], size['h']]
    # angle
    angle = rotatedRect.find('angle').attrib
    row.append(angle['d'])
    # points
    contour = space.find('contour')
    for point in contour.findall('point'):
        xy = point.attrib
        row = row + [xy['x'], xy['y']]
    csvwriter.writerow(row)
spaces_data.close()
```
